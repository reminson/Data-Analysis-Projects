```python
#For laptop model preparation
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

X_lp = df1.drop('Taken_product', axis=1)
y_lp = df1['Taken_product']

X_train_lp, X_test_lp, y_train_lp, y_test_lp = train_test_split(X_lp, y_lp, test_size=0.2, random_state=42)
```

```python
y_lp.head()
```

```python
# Initialize the Logistic Regression model
model_lp = LogisticRegression()

# Train the model on the training data
model_lp.fit(X_train_lp, y_train_lp)
```

```python
# Make predictions on the test data
predictions_LR = model_lp.predict(X_test_lp)

# Calculate accuracy
accuracy_lr = accuracy_score(y_test_lp, predictions_LR)
print(f'Accuracy: {accuracy_lr:.2f}')

# Print classification report
print(classification_report(y_test_lp, predictions_LR))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 100, random_state = 42, oob_score = True)
rf.fit(X_train_lp, y_train_lp)
```

```python
predictions_RF = rf.predict(X_test_lp).round()

# Calculate accuracy
accuracy_rf = accuracy_score(y_test_lp, predictions_RF)
print(f'Accuracy: {accuracy_rf:.2f}')

# Print classification report
print(classification_report(y_test_lp, predictions_RF))
```

```python
RF_lp_cm=confusion_matrix(y_test_lp, predictions_RF)
RF_lp_cm
```

```python
feature_importances_lp = rf.feature_importances_

# Create a DataFrame to store feature names and their importance values
feature_importance_df = pd.DataFrame({'Feature': X_lp.columns, 'Importance': feature_importances_lp})

# Sort the DataFrame by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Calculate feature importance percentages
feature_importance_df['Importance Percentage'] = (
    feature_importance_df['Importance'] / feature_importance_df['Importance'].sum() * 100
)

# Display the result
print("Feature Importance in Descending Order:")
print(feature_importance_df)
```

```python
from sklearn.svm import SVC
svc_lp = SVC(random_state=101)
svc_lp.fit(X_train_lp,y_train_lp)
```

```python
predictions_svc_lp = svc_lp.predict(X_test_lp).round()

# Calculate accuracy
accuracy_svc_lp = accuracy_score(y_test_lp, predictions_svc_lp)
print(f'Accuracy: {accuracy_svc_lp:.2f}')

# Print classification report
print(classification_report(y_test_lp, predictions_svc_lp))
```
